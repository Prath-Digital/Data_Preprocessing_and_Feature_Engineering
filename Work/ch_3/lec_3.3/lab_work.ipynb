```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import Binarizer, KBinsDiscretizer
np.random.seed(42)
```

# 🔹 Q1: Binning (Discretization)
### Convert continuous age into categories

```python
# Create Age column
ages = np.random.randint(18, 71, 100)
df_age = pd.DataFrame({'Age': ages})

# Define bins
bins = [18, 30, 50, 70]
labels = ['Young', 'Middle', 'Senior']

# Apply binning
df_age['Age_Group'] = pd.cut(df_age['Age'], bins=bins, labels=labels, right=True)

# Count values
counts = df_age['Age_Group'].value_counts()

df_age.head(), counts
```

# 🔹 Q2: Binarization
### Convert income into 0/1 based on threshold

```python
# Create Income data
income = np.random.randint(20000, 100001, 100)
df_income = pd.DataFrame({'Income': income})

# Apply Binarizer
binarizer = Binarizer(threshold=50000)
df_income['Income_Binary'] = binarizer.fit_transform(df_income[['Income']])

df_income.head()
```

# 🔹 Q3: Quantile Binning
### Split salary into equal-sized groups

```python
# Skewed salary data
salary = np.random.exponential(scale=50000, size=100)
df_salary = pd.DataFrame({'Salary': salary})

# Quantile binning
kbins_q = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='quantile')
df_salary['Quantile_Bin'] = kbins_q.fit_transform(df_salary[['Salary']])

# First 10 rows
df_salary.head(10)
```

```python
# Distribution count
df_salary['Quantile_Bin'].value_counts()
```

# 🔹 Q4: K-Means Binning
### Cluster-based binning

```python
# Spending Score
spending = np.random.randint(1, 101, 100)
df_spend = pd.DataFrame({'SpendingScore': spending})

# KMeans binning
kbins_k = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='kmeans')
df_spend['KMeans_Bin'] = kbins_k.fit_transform(df_spend[['SpendingScore']])

df_spend.head()
```

```python
# Compare with quantile
kbins_q2 = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='quantile')
df_spend['Quantile_Bin'] = kbins_q2.fit_transform(df_spend[['SpendingScore']])

df_spend.head()
```
