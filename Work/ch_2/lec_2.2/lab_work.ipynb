```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.impute import KNNImputer

np.random.seed(42)
```

# 🔹 Q1: Missing Indicator + Random Sample Imputation

## Step 1: Create Dataset with Missing Values

```python
data = {
    'Age': [25, 30, np.nan, 35, np.nan, 40, 28, np.nan],
    'Salary': [50000, 60000, 55000, np.nan, 65000, 70000, np.nan, 62000]
}

df = pd.DataFrame(data)
df
```

## Step 2: Add Missing Indicator Column

```python
df['Age_missing'] = df['Age'].isnull().astype(int)
df
```

## Step 3: Random Sample Imputation

```python
def random_impute(series):
    random_sample = series.dropna().sample(series.isnull().sum(), replace=True)
    random_sample.index = series[series.isnull()].index
    return series.fillna(random_sample)

df_imputed = df.copy()
df_imputed['Age'] = random_impute(df['Age'])
df_imputed
```

## Step 4: Compare Original vs Imputed

```python
print("Original Dataset:\n", df)
print("\nImputed Dataset:\n", df_imputed)
```

# 🔹 Q2: KNN Imputer (Multivariate Imputation)

## Step 1: Create Multivariate Dataset

```python
data2 = {
    'Age': [25, np.nan, 35, 40, np.nan, 29, 33, np.nan],
    'Salary': [50000, 60000, np.nan, 80000, 65000, np.nan, 72000, 69000],
    'Experience': [2, 5, 10, np.nan, 7, 3, np.nan, 6]
}

df2 = pd.DataFrame(data2)
df2
```

## Step 2: Apply KNN Imputer (k=3)

```python
imputer = KNNImputer(n_neighbors=3)
df2_imputed = pd.DataFrame(imputer.fit_transform(df2), columns=df2.columns)
df2_imputed
```

## Step 3: Compare Before and After

```python
print("Original Dataset:\n", df2)
print("\nImputed Dataset:\n", df2_imputed)
```

## Step 4: Experiment with Different K Values

```python
for k in [2, 3, 5]:
    imputer = KNNImputer(n_neighbors=k)
    temp = pd.DataFrame(imputer.fit_transform(df2), columns=df2.columns)
    print(f"\nK = {k}\n", temp)
```
