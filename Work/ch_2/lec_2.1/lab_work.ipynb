## Q1 - Missing Values Handling (Basic)

```python
import pandas as pd
import numpy as np

# Create sample dataset
data = {
    'Age': [22, np.nan, 24, 35, np.nan],
    'Fare': [7.25, 71.83, np.nan, 8.05, 10.5],
    'Gender': ['Male', 'Female', np.nan, 'Female', 'Male']
}

df = pd.DataFrame(data)
print('Original Dataset:')
print(df)

# Check missing values
print('\nMissing Values Count:')
print(df.isnull().sum())

# Drop rows with too many missing values
df_clean = df.dropna(thresh=2)
print('\nAfter Dropping Rows:')
print(df_clean)

# Fill missing values
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Fare'].fillna(df['Fare'].median(), inplace=True)
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)

print('\nAfter Imputation:')
print(df)
```

## Q2 - SimpleImputer (Numeric Data)

```python
from sklearn.impute import SimpleImputer

# Create numeric dataset
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, 4, np.nan]
}

df = pd.DataFrame(data)
print('Original Data:')
print(df)

# Mean Imputation
mean_imputer = SimpleImputer(strategy='mean')
df_mean = pd.DataFrame(mean_imputer.fit_transform(df), columns=df.columns)

print('\nAfter Mean Imputation:')
print(df_mean)

# Median Imputation
median_imputer = SimpleImputer(strategy='median')
df_median = pd.DataFrame(median_imputer.fit_transform(df), columns=df.columns)

print('\nAfter Median Imputation:')
print(df_median)
```

## Q3 - Categorical Imputation

```python
# Create categorical dataset
data = {
    'Gender': ['Male', 'Female', np.nan, 'Female', np.nan],
    'Department': ['HR', np.nan, 'IT', 'IT', np.nan]
}

df = pd.DataFrame(data)
print('Original Data:')
print(df)

# Most frequent (mode)
mode_imputer = SimpleImputer(strategy='most_frequent')
df_mode = pd.DataFrame(mode_imputer.fit_transform(df), columns=df.columns)

print('\nAfter Mode Imputation:')
print(df_mode)

# Constant replacement
constant_imputer = SimpleImputer(strategy='constant', fill_value='Unknown')
df_constant = pd.DataFrame(constant_imputer.fit_transform(df), columns=df.columns)

print('\nAfter Constant Imputation:')
print(df_constant)
```

## Q4 - Categorical Dataset Comparison

```python
# Dataset with Cities
data = {
    'City': ['Delhi', 'Mumbai', np.nan, np.nan, 'Delhi']
}

df = pd.DataFrame(data)
print('Original Data:')
print(df)

# Most Frequent
mf_imputer = SimpleImputer(strategy='most_frequent')
df_mf = pd.DataFrame(mf_imputer.fit_transform(df), columns=df.columns)

print('\nMost Frequent Imputation:')
print(df_mf)

# Constant
const_imputer = SimpleImputer(strategy='constant', fill_value='Unknown')
df_const = pd.DataFrame(const_imputer.fit_transform(df), columns=df.columns)

print('\nConstant Imputation:')
print(df_const)
```
