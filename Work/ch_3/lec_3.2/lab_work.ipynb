## 🔹 Q1: Ordinal Encoding
### Theory:
- Used when categories have natural order
- Example: Low < Medium < High

```python
import pandas as pd

# Create dataset
data = {'Education': ['High School', 'Graduate', 'Postgraduate', 'Graduate', 'High School']}
df = pd.DataFrame(data)
df
```

```python
# Define mapping
mapping = {'High School': 1, 'Graduate': 2, 'Postgraduate': 3}

# Apply mapping
df['Education_Encoded'] = df['Education'].map(mapping)
df
```

```python
# Verify order
df.sort_values('Education_Encoded')
```

## 🔹 Q2: Label Encoding
### Theory:
- Assigns unique number to each category
- No ordering meaning

```python
from sklearn.preprocessing import LabelEncoder

# Create dataset
data = {'City': ['Delhi', 'Mumbai', 'Chennai', 'Kolkata', 'Delhi']}
df2 = pd.DataFrame(data)

# Apply LabelEncoder
le = LabelEncoder()
df2['City_Encoded'] = le.fit_transform(df2['City'])
df2
```

```python
# Decode back
df2['City_Decoded'] = le.inverse_transform(df2['City_Encoded'])
df2
```

## 🔹 Q3: One-Hot Encoding
### Theory:
- Creates binary columns
- Removes ordering issue


```python
# Create dataset
data = {'Department': ['HR', 'IT', 'Finance', 'Sales', 'IT']}
df3 = pd.DataFrame(data)
df3
```

```python
# Using pandas get_dummies
pd.get_dummies(df3)
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

encoder = OneHotEncoder(sparse=False)
encoded = encoder.fit_transform(df3[['Department']])

# Convert to DataFrame
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
encoded_df
```

## ⚖️ Comparison

| Encoding Type | Best Use Case |
|--------------|--------------|
| Ordinal | When order matters |
| Label | Tree-based models |
| One-Hot | Linear models |

### Key Points:
- Linear models prefer One-Hot Encoding
- Tree models can work with Label Encoding
- Avoid Label Encoding for non-ordered data in linear models
