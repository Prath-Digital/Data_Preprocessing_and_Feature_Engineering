## Q3: Tensors (Pandas + PyTorch)

```python
import torch

# Scalar
scalar = torch.tensor(5)

# Vector
vector = torch.tensor([1, 2, 3])

# Matrix
matrix = torch.tensor([[1, 2], [3, 4]])

# 3D Tensor
tensor_3d = torch.tensor([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

# Print shapes
print("Scalar Shape:", scalar.shape)
print("Vector Shape:", vector.shape)
print("Matrix Shape:", matrix.shape)
print("3D Tensor Shape:", tensor_3d.shape)
```

## Q4: Tensor Operations

```python
import torch

# Generate random tensor (3,3,3)
tensor = torch.rand(3, 3, 3)
print("Original Tensor Shape:", tensor.shape)

# Reshape to (9,3)
reshaped = tensor.view(9, 3)
print("Reshaped Tensor Shape:", reshaped.shape)

# Slice first two layers
sliced = tensor[:2]
print("Sliced Tensor Shape:", sliced.shape)

# Matrix multiplication (2D tensors)
A = torch.rand(3, 3)
B = torch.rand(3, 3)
result = torch.matmul(A, B)
print("Matrix Multiplication Result Shape:", result.shape)

# Broadcasting
tensor1 = torch.rand(3, 1)
tensor2 = torch.rand(3, 3)
broadcast_result = tensor1 + tensor2

print("Broadcast Result Shape:", broadcast_result.shape)
```
