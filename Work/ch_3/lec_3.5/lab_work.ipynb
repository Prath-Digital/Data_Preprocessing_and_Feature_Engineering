## Q1: Feature Construction & Splitting

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. Split FullName into FirstName and LastName

```python
df = pd.DataFrame({
    'FullName': ['John Doe', 'Jane Smith', 'Alice Johnson']
})

df[['FirstName', 'LastName']] = df['FullName'].str.split(' ', expand=True)
df
```

### 2. Extract Year, Month, Day from OrderDate

```python
df['OrderDate'] = pd.to_datetime(['2023-01-15', '2022-12-20', '2024-03-10'])

df['Year'] = df['OrderDate'].dt.year
df['Month'] = df['OrderDate'].dt.month
df['Day'] = df['OrderDate'].dt.day

df
```

### 3. Create BMI Feature

```python
df['Height'] = [1.75, 1.60, 1.68]
df['Weight'] = [70, 60, 65]

df['BMI'] = df['Weight'] / (df['Height'] ** 2)
df
```

### 4. Word Count from Text Column

```python
df['Review'] = [
    'This product is amazing',
    'Not good',
    'Average quality but worth the price'
]

df['WordCount'] = df['Review'].apply(lambda x: len(x.split()))
df
```

# 📊 Q2: FunctionTransformer (Log, Reciprocal, Square Root)

```python
from sklearn.preprocessing import FunctionTransformer
```

### 1. Log Transformation on Income

```python
income = np.random.randint(10000, 200000, 1000).reshape(-1, 1)

log_transformer = FunctionTransformer(np.log1p)
income_log = log_transformer.fit_transform(income)

plt.figure()
sns.histplot(income, kde=True)
plt.title('Original Income Distribution')

plt.figure()
sns.histplot(income_log, kde=True)
plt.title('Log Transformed Income')
```

### 2. Reciprocal Transformation on Age

```python
age = np.random.randint(18, 60, 1000).reshape(-1, 1)

reciprocal_transformer = FunctionTransformer(lambda x: 1 / x)
age_reciprocal = reciprocal_transformer.fit_transform(age)

plt.figure()
sns.histplot(age, kde=True)
plt.title('Original Age')

plt.figure()
sns.histplot(age_reciprocal, kde=True)
plt.title('Reciprocal Transformed Age')
```

### 3. Square Root Transformation on PurchaseCount

```python
purchase = np.random.randint(1, 100, 1000).reshape(-1, 1)

sqrt_transformer = FunctionTransformer(np.sqrt)
purchase_sqrt = sqrt_transformer.fit_transform(purchase)

plt.figure()
sns.histplot(purchase, kde=True)
plt.title('Original PurchaseCount')

plt.figure()
sns.histplot(purchase_sqrt, kde=True)
plt.title('Square Root Transformed PurchaseCount')
```

### 4. Compare Skewness

```python
import scipy.stats as stats

print('Income Skewness Before:', stats.skew(income.flatten()))
print('Income Skewness After Log:', stats.skew(income_log.flatten()))

print('Purchase Skewness Before:', stats.skew(purchase.flatten()))
print('Purchase Skewness After Sqrt:', stats.skew(purchase_sqrt.flatten()))
```
