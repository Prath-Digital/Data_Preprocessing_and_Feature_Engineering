## Q1 - Working with CSV File

```python
import pandas as pd

# Create sample CSV
data = {
    'Name': ['A', 'B', 'C', 'D', 'E'],
    'Age': [20, 21, 19, 22, 20],
    'Marks': [75, 60, 85, 90, 70]
}
df = pd.DataFrame(data)
df.to_csv('students.csv', index=False)

# Load CSV
df = pd.read_csv('students.csv')

# Display info
print(df.head())
print(df.columns)
print(df.shape)

# Filter marks > 70
filtered_df = df[df['Marks'] > 70]

# Save new CSV
filtered_df.to_csv('filtered_students.csv', index=False)
filtered_df
```

## Q2 - Working with JSON File

```python
import json

# Create JSON data
employees = [
    {'id': 1, 'name': 'John', 'department': 'HR'},
    {'id': 2, 'name': 'Anna', 'department': 'IT'},
    {'id': 3, 'name': 'Mike', 'department': 'Finance'}
]

with open('employees.json', 'w') as f:
    json.dump(employees, f)

# Load JSON into pandas
df_json = pd.read_json('employees.json')
df_json

# Convert back to JSON
df_json.to_json('output.json', orient='records', indent=4)
```

## Q5 - Data Cleaning Process

```python
# Create dataset with missing values
import numpy as np

data_clean = {
    'A': [1, 2, np.nan, 4],
    'B': [5, np.nan, np.nan, 8],
    'C': [10, 20, 30, 40]
}

df_clean = pd.DataFrame(data_clean)

# Info & describe
print(df_clean.info())
print(df_clean.describe())

# Fill missing values
df_clean['A'].fillna(df_clean['A'].mean(), inplace=True)
df_clean['B'].fillna(df_clean['B'].median(), inplace=True)

# Drop duplicates
df_clean.drop_duplicates(inplace=True)

# Min-Max Normalization
df_clean['C'] = (df_clean['C'] - df_clean['C'].min()) / (df_clean['C'].max() - df_clean['C'].min())

df_clean
```
