```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, Normalizer, MinMaxScaler, MaxAbsScaler, RobustScaler
```

## 🔹 Q1: Standardization

```python
# Create DataFrame
df = pd.DataFrame({
    'Height': [150, 160, 170, 180, 190],
    'Weight': [50, 60, 70, 80, 90]
})

print("Before Scaling:")
print(df.mean(), df.std())

# Apply StandardScaler
scaler = StandardScaler()
scaled = scaler.fit_transform(df)
df_scaled = pd.DataFrame(scaled, columns=df.columns)

print("\nAfter Scaling:")
print(df_scaled.mean(), df_scaled.std())
```

## 🔹 Q2: Normalization (L2)

```python
# Dataset
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
})

# Normalize
normalizer = Normalizer(norm='l2')
normalized = normalizer.fit_transform(df)

df_norm = pd.DataFrame(normalized, columns=df.columns)
print(df_norm)

# Verify norm = 1
norms = np.linalg.norm(df_norm, axis=1)
print("\nRow norms:", norms)
```

## 🔹 Q3: Min-Max Scaling

```python
# Salary Data
salary = np.linspace(20000, 100000, 100)
df = pd.DataFrame({'Salary': salary})

# Before Plot
plt.hist(df['Salary'])
plt.title("Before Scaling")
plt.show()

# Scaling
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df)
df_scaled = pd.DataFrame(scaled, columns=['Salary'])

# After Plot
plt.hist(df_scaled['Salary'])
plt.title("After Scaling")
plt.show()
```

## 🔹 Q4: MaxAbs Scaling

```python
# Random Data
data = np.random.randint(-100, 100, size=(10, 1))
df = pd.DataFrame(data, columns=['Values'])

print("Before:\n", df.describe())

# Scaling
scaler = MaxAbsScaler()
scaled = scaler.fit_transform(df)
df_scaled = pd.DataFrame(scaled, columns=['Values'])

print("\nAfter:\n", df_scaled.describe())
```

## 🔹 Q5: Robust Scaling

```python
# Data with Outliers
data = np.array([10, 12, 14, 15, 18, 1000])
df = pd.DataFrame({'Income': data})

# StandardScaler
std = StandardScaler()
df_std = std.fit_transform(df)

# RobustScaler
robust = RobustScaler()
df_robust = robust.fit_transform(df)

# Plot comparison
plt.boxplot([df_std.flatten(), df_robust.flatten()], labels=['Standard', 'Robust'])
plt.title("Comparison")
plt.show()
```
